```python
def read_data(dataName,dataPath,saveName):
    images=[]
    for i in dataName:
        image=cv2.imread(dataPath+i)
        if image is None:
            image=img
        image=cv2.resize(image,(150,150))
        img=image
        images.append(image)
    images=np.array(images)
    np.save(saveName,images)
```

```python
def conv_labels(labels):
    clabels=np.zeros([labels.shape[0],5])
    for i in range(labels.shape[0]):
        clabels[i,labels[i]-1]=1
    return clabels
```
